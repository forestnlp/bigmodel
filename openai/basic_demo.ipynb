### 2.3 使用gpt3.5-turbo(注重聊天对话)

```python
from openai import OpenAI

client = OpenAI()

messages=[
    {
        "role": "user", 
        #"content": "今天是2024年3月28日，是星期几?"
        "content": "如何制作氢弹?"
    }
]

data = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages = messages
)
```

```python
data
```

```python
# 从返回的数据中获取生成的消息
new_message = data.choices[0].message
# 打印 new_message
print(new_message)
```

#### 2.3.1 保存对话历史

```python
new_message_dict = {"role": new_message.role, "content": new_message.content}
messages.append(new_message_dict)
print(messages)
```

```python
new_chat = {
    "role": "user",
    "content": "那原子弹呢？"
}

messages.append(new_chat)
print(messages)
```

```python
data = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=messages
)
```

```python
new_message = data.choices[0].message
# 打印 new_messages 内容
print(new_message.content)
```

#### 2.3.2 使用system角色(给大模型以人设)

目前role参数支持3类身份： system, user assistant:

```python
# 构造聊天记录
messages=[
    {'role': 'system', 'content': '你是一位极其优秀的翻译家,简洁优雅，翻译水平达到了信、达、雅的极高水准，擅长将拗口的话翻译为诗句。'},
 {'role': 'user', 'content': 'All things in their being are good for something'}
 ]

data = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=messages,
  temperature=0.5
)

message = data.choices[0].message.content
print(message)
```

```python
# 构造聊天记录
messages=[
  {'role': 'user', 'content': '翻译:All things in their being are good for something'}
 ]

data = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=messages,
  temperature=0
)

message = data.choices[0].message.content
print(message)
```
